```python
from tkinter import *
```

```python
root = Tk()
root.title("CALCULATER")

e = Entry(root, width = 60, borderwidth = 10)
e.grid(row = 0 , column = 0 , columnspan = 3, padx = 10, pady = 10)

def button_click(number):
    current = e.get()
    e.delete(0,END)
    e.insert(0,str(current) + str(number))
    
def button_clear():
    e.delete(0,END)
    
def button_add():
    first_num = e.get()
    global f_num
    global math
    math = "add"
    f_num = int(first_num)
    e.delete(0,END)
    
def button_equal():
    second_num = e.get()
    e.delete (0,END)
    if math == "add":
        e.insert(0,f_num+int(second_num))
    elif math == "substraction":
        
        e.insert(0,f_num - int(second_num))
        
    elif math == "multiplication":
        e.insert(0,f_num*int(second_num))
    else:
         e.insert(0,f_num/int(second_num))
    
    
def button_subs():
    first_num = e.get()
    global f_num
    global math
    math = "substraction"
    f_num = int(first_num)
    e.delete(0,END)

    
def  button_multi():  
    first_num = e.get()
    global f_num
    global math
    math = "multiplication"
    f_num = int(first_num)
    e.delete(0,END)
    
def  button_div():
    first_num = e.get()
    global f_num
    global math
    math = "division"
    f_num = int(first_num)
    e.delete(0,END) 
    
    
button_1 = Button(root, text= "1",padx = 40,pady = 22, command = lambda: button_click(1),fg = "black",bg = "skyblue").grid(row = 3,column =0)
button_2 = Button(root, text= "2",padx =  40,pady = 22, command = lambda: button_click(2),fg = "black",bg = "skyblue").grid(row = 3,column =1)
button_3 = Button(root, text= "3",padx = 40,pady = 22, command = lambda:button_click(3),fg = "black",bg = "skyblue").grid(row = 3,column =2)
button_4 = Button(root, text= "4",padx =  40,pady = 22, command = lambda:  button_click(4),fg = "black",bg = "skyblue").grid(row = 2,column =0)
button_5 = Button(root, text= "5",padx =  40,pady = 22, command = lambda: button_click(5),fg = "black",bg = "skyblue").grid(row = 2,column =1)
button_6 = Button(root, text= "6",padx =  40,pady = 22, command =lambda: button_click(6),fg = "black",bg = "skyblue").grid(row = 2,column =2)
button_7 = Button(root, text= "7",padx =  40,pady = 22, command =lambda: button_click(7),fg = "black",bg = "skyblue").grid(row = 1,column =0)
button_8 = Button(root, text= "8",padx =  40,pady = 22, command =lambda:  button_click(8),fg = "black",bg = "skyblue").grid(row = 1,column =1)
button_9 = Button(root, text= "9",padx =  40,pady = 22, command = lambda: button_click(9),fg = "black",bg = "skyblue").grid(row = 1,column =2)
button_0 = Button(root, text= "0",padx =  40,pady = 22, command = lambda: button_click(0),fg = "black",bg = "skyblue").grid(row = 1,column =3 )
button_10 = Button(root,text= "=",padx = 40,pady = 22,command =  button_equal,fg = "black",bg = "skyblue").grid(row = 2,columnspan = 3, column =3 )
button_11 =Button(root, text="AC",padx = 40,pady = 22,command =button_clear,fg = "black",bg = "skyblue").grid(row = 3,columnspan = 3, column =3 )
button_12 = Button(root, text= "+",padx = 40,pady = 22, command = button_add,fg = "black",bg = "skyblue").grid(row = 4,column =0 )
button_13= Button(root, text= "-",padx = 40,pady = 22, command = button_subs,fg = "black",bg = "skyblue").grid(row = 4,column =1)
button_14= Button(root, text= "*",padx = 40,pady = 22, command =  button_multi,fg = "black",bg = "skyblue").grid(row = 4,column =2 )
button_15= Button(root, text= "/",padx = 40,pady = 22, command =  button_div,fg = "black",bg = "skyblue").grid(row = 4,column =3 )



root.mainloop()



```
